# Scatter plot game

```python
import matplotlib.pyplot as plt
import random
import numpy as np

class ScatterPlotGame:
    def __init__(self):
        self.points = []
        self.current_point_index = 0
        self.score = 0
        self.difficulty = "medium"  # Default difficulty

    def set_difficulty(self):
        """Allow player to choose difficulty level"""
        print("\nChoose difficulty level:")
        print("1. Easy (Graph: -10 to 10)")
        print("2. Medium (Graph: -20 to 20)")
        print("3. Hard (Graph: -50 to 50)")
        print("4. Expert (Graph: -100 to 100)")

        choice = input("Enter your choice (1-4): ").strip()
        difficulty_map = {
            "1": ("easy", 10),
            "2": ("medium", 20),
            "3": ("hard", 50),
            "4": ("expert", 100)
        }

        if choice in difficulty_map:
            self.difficulty, self.graph_range = difficulty_map[choice]
            print(f"\nDifficulty set to {self.difficulty.capitalize()}!")
        else:
            print("Invalid choice. Defaulting to Medium difficulty.")
            self.difficulty, self.graph_range = "medium", 20

    def generate_points(self, num_points=5):
        """Generate random points based on difficulty"""
        self.points = []
        for _ in range(num_points):
            x = random.randint(-self.graph_range, self.graph_range)
            y = random.randint(-self.graph_range, self.graph_range)
            self.points.append((x, y))

    def display_graph(self):
        """Display the scatter plot with numbered points"""
        plt.figure(figsize=(8, 8))

        # Extract x and y coordinates
        x_coords = [point[0] for point in self.points]
        y_coords = [point[1] for point in self.points]

        # Create scatter plot
        plt.scatter(x_coords, y_coords, color='red', s=100, zorder=5)

        # Add numbers to each point
        for i, (x, y) in enumerate(self.points):
            plt.annotate(f"{i+1}", (x, y), xytext=(5, 5), textcoords='offset points',
                        fontsize=12, fontweight='bold', color='darkred', zorder=6)

        # Set graph limits and style
        plt.xlim(-self.graph_range-5, self.graph_range+5)
        plt.ylim(-self.graph_range-5, self.graph_range+5)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.axhline(y=0, color='black', linewidth=1)
        plt.axvline(x=0, color='black', linewidth=1)

        # Add labels and title
        plt.title(f"Scatter Plot Game - {self.difficulty.capitalize()} Mode", fontsize=14)
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")

        # Show the plot
        plt.tight_layout()
        plt.show()

    def play_round(self):
        """Play one round of the game"""
        if self.current_point_index >= len(self.points):
            return False

        point_number = self.current_point_index + 1
        print(f"\nPoint {point_number} of {len(self.points)}")
        print(f"Enter the coordinates for Point #{point_number}:")

        try:
            x_guess = int(input("Enter the X coordinate: "))
            y_guess = int(input("Enter the Y coordinate: "))
        except ValueError:
            print("Please enter valid integers for coordinates.")
            return True

        actual_x, actual_y = self.points[self.current_point_index]

        if x_guess == actual_x and y_guess == actual_y:
            print("✅ Correct! Well done!")
            self.score += 1
        else:
            print(f"❌ Incorrect. The correct coordinates were ({actual_x}, {actual_y})")

        self.current_point_index += 1
        return True

    def show_final_score(self):
        """Display the final score and correct answers"""
        print("\n" + "="*50)
        print("GAME OVER!")
        print(f"Your final score: {self.score}/{len(self.points)}")

        if self.score == len(self.points):
            print("🎉 Perfect score! You're a coordinate master!")
        elif self.score >= len(self.points) * 0.7:
            print("👍 Great job! You have a good eye for coordinates!")
        elif self.score >= len(self.points) * 0.5:
            print("😊 Not bad! Keep practicing!")
        else:
            print("💪 Keep practicing! You'll get better!")

        print("\nThe correct coordinates were:")
        for i, (x, y) in enumerate(self.points):
            print(f"Point {i+1}: ({x}, {y})")

    def play_game(self):
        """Main game loop"""
        print("="*50)
        print("Welcome to the Scatter Plot Game!")
        print("Identify the coordinates of points on the graph.")
        print("="*50)

        # Set difficulty
        self.set_difficulty()

        # Generate points
        num_points = 5
        if self.difficulty == "hard":
            num_points = 7
        elif self.difficulty == "expert":
            num_points = 10

        self.generate_points(num_points)

        # Show the graph
        print("\nGet ready! The graph will appear in a moment...")
        print("Each point is numbered to help you identify them.")
        input("Press Enter to continue...")
        self.display_graph()

        # Play rounds
        playing = True
        while playing:
            playing = self.play_round()

        # Show final results
        self.show_final_score()

        # Ask if player wants to play again
        play_again = input("\nWould you like to play again? (y/n): ").strip().lower()
        if play_again == 'y' or play_again == 'yes':
            self.__init__()  # Reset the game
            self.play_game()
        else:
            print("Thanks for playing!")

# Run the game
if __name__ == "__main__":
    game = ScatterPlotGame()
    game.play_game()
```

# Algebra practice game

```python
import random
import time

class AlgebraGame:
    def __init__(self):
        self.score = 0
        self.total_questions = 0
        self.difficulty = "medium"
        self.number_range = 20  # Default range

    def set_difficulty(self):
        """Set the difficulty level which affects number range"""
        print("\nChoose difficulty level:")
        print("1. Easy (Numbers: -10 to 10)")
        print("2. Medium (Numbers: -20 to 20)")
        print("3. Hard (Numbers: -50 to 50)")
        print("4. Expert (Numbers: -100 to 100)")

        choice = input("Enter your choice (1-4): ").strip()
        difficulty_map = {
            "1": ("easy", 10),
            "2": ("medium", 20),
            "3": ("hard", 50),
            "4": ("expert", 100)
        }

        if choice in difficulty_map:
            self.difficulty, self.number_range = difficulty_map[choice]
            print(f"\nDifficulty set to {self.difficulty.capitalize()}!")
        else:
            print("Invalid choice. Defaulting to Medium difficulty.")
            self.difficulty, self.number_range = "medium", 20

    def generate_one_step_equation(self):
        """Generate a one-step equation: ax + b = c"""
        a = random.randint(1, self.number_range // 2)
        x = random.randint(-self.number_range, self.number_range)
        b = random.randint(-self.number_range, self.number_range)

        # Calculate c such that ax + b = c
        c = a * x + b

        # Randomly choose equation format
        equation_type = random.choice([
            "addition", "subtraction", "multiplication", "division"
        ])

        if equation_type == "addition":
            # x + b = c
            return f"x + {b} = {c}", x
        elif equation_type == "subtraction":
            # x - b = c
            return f"x - {b} = {c}", x
        elif equation_type == "multiplication":
            # ax = c
            return f"{a}x = {a * x}", x
        else:  # division
            # x/a = b (but we'll present it as multiplication to avoid fractions)
            # Instead, let's do: x ÷ a = b where b is integer
            a = random.randint(2, self.number_range // 4)  # Avoid division by 1
            x = random.randint(-self.number_range, self.number_range)
            # Make sure x is divisible by a
            x = a * random.randint(-self.number_range // a, self.number_range // a)
            return f"x ÷ {a} = {x // a}", x

    def generate_two_step_equation(self):
        """Generate a two-step equation: ax + b = c"""
        a = random.randint(2, self.number_range // 4)
        x = random.randint(-self.number_range, self.number_range)
        b = random.randint(-self.number_range, self.number_range)
        c = a * x + b

        # Randomly choose the format
        format_type = random.choice([1, 2])

        if format_type == 1:
            return f"{a}x + {b} = {c}", x
        else:
            return f"{b} + {a}x = {c}", x

    def generate_equation_with_parentheses(self):
        """Generate equation with parentheses: a(bx + c) = d"""
        a = random.randint(2, 5)
        b = random.randint(1, 3)
        x = random.randint(-self.number_range // 2, self.number_range // 2)
        c = random.randint(-self.number_range // 2, self.number_range // 2)

        d = a * (b * x + c)
        return f"{a}({b}x + {c}) = {d}", x

    def format_negative_numbers(self, text):
        """Format negative numbers to show parentheses for better readability"""
        return text.replace("-", "–")

    def ask_question(self, question_number):
        """Ask a single algebra question"""
        # Randomly choose question type based on difficulty
        if self.difficulty == "easy":
            question_types = ["one_step"] * 3 + ["two_step"]
        elif self.difficulty == "medium":
            question_types = ["one_step"] * 2 + ["two_step"] * 2
        else:  # hard or expert
            question_types = ["one_step", "two_step", "parentheses"]

        question_type = random.choice(question_types)

        if question_type == "one_step":
            equation, answer = self.generate_one_step_equation()
        elif question_type == "two_step":
            equation, answer = self.generate_two_step_equation()
        else:  # parentheses
            equation, answer = self.generate_equation_with_parentheses()

        print(f"\nQuestion {question_number}:")
        print(f"Solve for x: {self.format_negative_numbers(equation)}")

        start_time = time.time()

        try:
            user_answer = float(input("x = ").strip())
        except ValueError:
            print("❌ Please enter a valid number!")
            return False

        answer_time = time.time() - start_time

        # Allow small rounding error for floating point
        if abs(user_answer - answer) < 0.001:
            time_bonus = max(0, 1 - answer_time / 10)  # Bonus for quick answers
            points = 1 + time_bonus
            self.score += points
            print(f"✅ Correct! (+{points:.1f} points)")
            if answer_time < 3:
                print("⚡ Quick thinking!")
            return True
        else:
            print(f"❌ Incorrect. The answer was x = {answer}")
            return False

    def show_hint(self):
        """Show helpful algebra hints"""
        hints = [
            "\n💡 Hint: For x + a = b, subtract a from both sides: x = b - a",
            "💡 Hint: For x - a = b, add a to both sides: x = b + a",
            "💡 Hint: For ax = b, divide both sides by a: x = b ÷ a",
            "💡 Hint: For x ÷ a = b, multiply both sides by a: x = b × a",
            "💡 Hint: For two-step equations, undo addition/subtraction first, then multiplication/division",
            "💡 Hint: For a(bx + c) = d, first divide by a, then solve the remaining equation",
            "💡 Hint: Watch out for negative signs! Double-check your arithmetic.",
            "💡 Hint: Take your time and write down each step if needed."
        ]
        print(random.choice(hints))

    def play_game(self):
        """Main game loop"""
        print("="*60)
        print("Welcome to the Algebra Practice Game!")
        print("Solve equations for x. Use integers or decimals in your answers.")
        print("="*60)

        self.set_difficulty()

        # Determine number of questions based on difficulty
        if self.difficulty == "easy":
            num_questions = 8
        elif self.difficulty == "medium":
            num_questions = 10
        elif self.difficulty == "hard":
            num_questions = 12
        else:  # expert
            num_questions = 15

        print(f"\nYou'll be solving {num_questions} algebra problems.")
        print("Type 'hint' for a helpful tip during any question!")
        input("\nPress Enter to start...")

        for i in range(num_questions):
            self.total_questions += 1

            # Allow user to ask for hint
            if i > 0 and i % 3 == 0:  # Offer hint every 3 questions
                hint_choice = input("\nWould you like a hint before the next question? (y/n): ").lower()
                if hint_choice == 'y':
                    self.show_hint()

            success = self.ask_question(i + 1)

            if not success and random.random() < 0.3:  # 30% chance after wrong answer
                self.show_hint()

        self.show_final_results()

        # Ask if player wants to play again
        play_again = input("\nWould you like to play again? (y/n): ").strip().lower()
        if play_again == 'y' or play_again == 'yes':
            self.__init__()
            self.play_game()
        else:
            print("Thanks for practicing algebra!")

    def show_final_results(self):
        """Display final score and performance evaluation"""
        percentage = (self.score / self.total_questions) * 100

        print("\n" + "="*50)
        print("GAME OVER - FINAL RESULTS")
        print("="*50)
        print(f"Final Score: {self.score:.1f}/{self.total_questions} ({percentage:.1f}%)")

        if percentage >= 90:
            print("🎉 Algebra Master! Perfect work!")
            print("You're ready for more advanced math!")
        elif percentage >= 80:
            print("🌟 Excellent! You have strong algebra skills!")
        elif percentage >= 70:
            print("👍 Good job! You understand the concepts well!")
        elif percentage >= 60:
            print("😊 Not bad! Keep practicing to improve!")
        else:
            print("💪 Keep working at it! Algebra takes practice!")
            print("Remember to review the steps for solving equations.")

        # Show study recommendations
        if percentage < 70:
            print("\n📚 Study Recommendation:")
            if self.difficulty == "expert" or self.difficulty == "hard":
                print("Try practicing with Medium difficulty to build confidence.")
            print("Focus on one-step equations first, then move to two-step equations.")

# Run the game
if __name__ == "__main__":
    game = AlgebraGame()
    game.play_game()
```

# Projectile game

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, FloatText, Button, Output, VBox, HBox
import random
from IPython.display import display, clear_output

class ProjectileGame:
    def __init__(self):
        self.wall_x = 0
        self.wall_height = 0
        self.level = 1
        self.score = 0
        self.main_output = Output()

    def generate_wall(self):
        """Generate a random wall position and height"""
        self.wall_x = random.uniform(3, 8)
        self.wall_height = random.uniform(1, 4)

    def parabola(self, x, a, b, c):
        """Calculate y value for parabolic path: y = ax² + bx + c"""
        return a*x**2 + b*x + c

    def check_clearance(self, a, b, c):
        """Check if the projectile clears the wall"""
        wall_y = self.parabola(self.wall_x, a, b, c)
        return wall_y > self.wall_height

    def plot_game(self, a, b, c, show_result=False):
        """Plot the game scene with projectile path and wall"""
        plt.figure(figsize=(12, 6))

        # Generate x values for plotting
        x = np.linspace(0, 10, 400)
        y = self.parabola(x, a, b, c)

        # Plot the parabolic path
        plt.plot(x, y, 'b-', linewidth=2, label=f'y = {a:.2f}x² + {b:.2f}x + {c:.2f}')

        # Plot the wall
        wall_width = 0.2
        plt.fill_between([self.wall_x - wall_width/2, self.wall_x + wall_width/2],
                        0, self.wall_height, color='red', alpha=0.7, label='Wall')

        # Plot ground
        plt.axhline(y=0, color='green', linewidth=3, label='Ground')

        # Customize plot
        plt.xlim(0, 10)
        plt.ylim(0, 8)
        plt.xlabel('Distance')
        plt.ylabel('Height')
        plt.title(f'Projectile Game - Level {self.level} | Score: {self.score}')
        plt.grid(True, alpha=0.3)
        plt.legend()

        # Show result if checking
        if show_result:
            if self.check_clearance(a, b, c):
                plt.text(5, 7, 'SUCCESS! 🎯', fontsize=20, ha='center',
                        bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen"))
            else:
                plt.text(5, 7, 'FAILED! 💥', fontsize=20, ha='center',
                        bbox=dict(boxstyle="round,pad=0.3", facecolor="lightcoral"))

        plt.show()

    def run_level_1(self):
        """Run level 1 with proper waiting"""
        with self.main_output:
            clear_output(wait=True)
            print("🎯 LEVEL 1: Use sliders to adjust the parabolic path!")
            print(f"Wall at x={self.wall_x:.1f}, height={self.wall_height:.1f}")
            print("Adjust a, b, c sliders, then click 'Launch Projectile!'")
            print("Click 'Next Level' when you're ready to move on.")

        # Create sliders
        a_slider = FloatSlider(value=0.0, min=-2.0, max=2.0, step=0.1,
                              description='a (x² coefficient):', style={'description_width': 'initial'})
        b_slider = FloatSlider(value=2.0, min=-5.0, max=5.0, step=0.1,
                              description='b (x coefficient):', style={'description_width': 'initial'})
        c_slider = FloatSlider(value=0.0, min=-2.0, max=5.0, step=0.1,
                              description='c (constant):', style={'description_width': 'initial'})

        # Create buttons
        launch_button = Button(description="Launch Projectile!", button_style='success')
        next_level_button = Button(description="Next Level →", button_style='info')

        # Create outputs
        preview_output = Output()
        result_output = Output()
        level_complete_output = Output()

        def update_preview(a, b, c):
            """Update the preview plot when sliders change"""
            with preview_output:
                preview_output.clear_output(wait=True)
                self.plot_game(a, b, c, show_result=False)

        def on_slider_change(change):
            """Update preview when any slider changes"""
            a = a_slider.value
            b = b_slider.value
            c = c_slider.value
            update_preview(a, b, c)

        def on_launch_click(b):
            """Check results when launch button is clicked"""
            a = a_slider.value
            b_val = b_slider.value
            c = c_slider.value

            with result_output:
                result_output.clear_output(wait=True)
                # Show the final graph with result
                self.plot_game(a, b_val, c, show_result=True)

                # Also show text feedback
                if self.check_clearance(a, b_val, c):
                    print("🎯 SUCCESS! You cleared the wall!")
                    print(f"Height at wall: {self.parabola(self.wall_x, a, b_val, c):.2f} (needed > {self.wall_height:.2f})")
                    self.score += 1
                else:
                    height_at_wall = self.parabola(self.wall_x, a, b_val, c)
                    print(f"💥 FAILED! Height at wall: {height_at_wall:.2f} (needed > {self.wall_height:.2f})")
                    print("Keep adjusting the sliders and try again!")

        def on_next_level_click(b):
            """Move to level 2 when next level button is clicked"""
            with level_complete_output:
                level_complete_output.clear_output(wait=True)
                print("Moving to Level 2...")
                self.run_level_2()

        # Connect events
        a_slider.observe(on_slider_change, names='value')
        b_slider.observe(on_slider_change, names='value')
        c_slider.observe(on_slider_change, names='value')
        launch_button.on_click(on_launch_click)
        next_level_button.on_click(on_next_level_click)

        # Initial preview
        update_preview(a_slider.value, b_slider.value, c_slider.value)

        # Display widgets
        display(VBox([
            self.main_output,
            a_slider,
            b_slider,
            c_slider,
            HBox([launch_button, next_level_button]),
            preview_output,
            result_output,
            level_complete_output
        ]))

    def run_level_2(self):
        """Run level 2"""
        self.level = 2
        self.generate_wall()

        with self.main_output:
            clear_output(wait=True)
            print("🎯 LEVEL 2: Enter a, b, c values directly!")
            print(f"Wall at x={self.wall_x:.1f}, height={self.wall_height:.1f}")
            print("This is harder - no sliders to guide you!")
            print("Click 'Finish Game' when you're done.")

        # Create input fields
        a_input = FloatText(value=0.0, description='a:', style={'description_width': 'initial'})
        b_input = FloatText(value=2.0, description='b:', style={'description_width': 'initial'})
        c_input = FloatText(value=0.0, description='c:', style={'description_width': 'initial'})

        # Create buttons
        check_button = Button(description="Launch Projectile!", button_style='success')
        finish_button = Button(description="Finish Game", button_style='warning')

        # Create outputs
        result_output = Output()
        game_complete_output = Output()

        def on_check_click(b):
            a = a_input.value
            b_val = b_input.value
            c = c_input.value

            with result_output:
                result_output.clear_output(wait=True)
                self.plot_game(a, b_val, c, show_result=True)

                if self.check_clearance(a, b_val, c):
                    print("🎯 Excellent! Direct hit!")
                    self.score += 2  # Bonus points for level 2
                else:
                    height_at_wall = self.parabola(self.wall_x, a, b_val, c)
                    print(f"💥 Crash! Height at wall: {height_at_wall:.2f} (needed > {self.wall_height:.2f})")

        def on_finish_click(b):
            """Show final results when finish button is clicked"""
            with game_complete_output:
                game_complete_output.clear_output(wait=True)
                self.show_final_results()

        check_button.on_click(on_check_click)
        finish_button.on_click(on_finish_click)

        display(VBox([
            self.main_output,
            a_input,
            b_input,
            c_input,
            HBox([check_button, finish_button]),
            result_output,
            game_complete_output
        ]))

    def show_final_results(self):
        """Display final game results"""
        with self.main_output:
            clear_output(wait=True)
            print("🎮 GAME COMPLETE!")
            print("=" * 30)
            print(f"FINAL SCORE: {self.score}")
            print("=" * 30)

            if self.score >= 3:
                print("🏆 CHAMPION! Perfect score!")
            elif self.score >= 1:
                print("👍 Good job!")
            else:
                print("💪 Keep practicing!")

    def start_game(self):
        """Start the projectile game"""
        with self.main_output:
            print("🚀 PROJECTILE GAME")
            print("=" * 50)
            print("Goal: Adjust the parabolic path to clear the wall!")
            print("Level 1: Use sliders to find the right parameters")
            print("Level 2: Enter a, b, c values directly (harder!)")
            print("=" * 50)

        # Start level 1
        self.generate_wall()
        self.run_level_1()

# Run the game
def start_projectile_game():
    """Start the projectile game"""
    game = ProjectileGame()
    game.start_game()

# Start the game
start_projectile_game()
```
