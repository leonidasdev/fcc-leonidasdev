# Step 8 - Making a Scatterplot Game

To make the game, you can make a loop that plots a random point and asks the user to input the (x,y) coordinates. Notice the `for` loop that runs three rounds of the game. Run the code, play the game, then you can go on to the next step.

```python
import matplotlib.pyplot as plt
import random

score = 0

xmin = -8
xmax = 8
ymin = -8
ymax = 8

fig, ax = plt.subplots()

for i in range(0,3):
    xpoint = random.randint(xmin, xmax)
    ypoint = random.randint(ymin, ymax)
    x = [xpoint]
    y = [ypoint]
    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'b') # blue x axis
    plt.plot([0,0],[ymin,ymax], 'b') # blue y axis
    plt.plot(x, y, 'ro')
    print(" ")
    plt.grid() # displays grid lines on graph
    plt.show()
    guess = input("Enter the coordinates of the red point point: \n")
    guess_array = guess.split(",")
    xguess = int(guess_array[0])
    yguess = int(guess_array[1])
    if xguess == xpoint and yguess == ypoint:
        score = score + 1

print("Your score: ", score) # notice this is not in the loop


# Only change code above this line
import math_code_test_b as test
test.step08(score)
```

# Step 13 - Solutions as Coordinates

The `linsolve()` function returns a <i>finite set</i>, and you can convert that "finite set" into (x,y) coordinates. Notice how the code parses the `solution` variable into two separate variables. Just run the code to see how this works.

```python
from sympy import *
x,y = symbols('x y')

# Use variables for each equation
first = x + y
second = x - y

# parse finite set answer as coordinate pair
solution = linsolve([first, second], (x, y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]

print("x = ", x_solution)
print("y = ", y_solution)
print(" ")
print("Solution: (",x_solution,",",y_solution,")")


# Just run this code
import math_code_test_b as test
test.step01()
```

# Step 15 - Solve and graph a system

Now you can put it all together: solve a system of equations, graph the system, and plot a point where the two lines intersect. Notice how this code is not like the previous solving equations code or the graphing code or the user input code. Python uses `sympy` to get the (x,y) solution and `numpy` to get the values to graph, so the user inputs nummerical values and the code uses them in two different ways. Think about how you would do this if the user input values for ax + by = c.

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

print("First equation: y = mx + b")
mb_1 = input("Enter m and b, separated by a comma: ")
mb_in1 = mb_1.split(",")
m1 = float(mb_in1[0])
b1 = float(mb_in1[1])

print("Second equation: y = mx + b")
mb_2 = input("Enter m and b, separated by a comma: ")
mb_in2 = mb_2.split(",")
m2 = float(mb_in2[0])
b2 = float(mb_in2[1])

# Solve the system of equations
x,y = symbols('x y')
first = m1*x + b1 - y
second = m2*x + b2 - y
solution = linsolve([first, second], (x, y))
x_solution = round(float(solution.args[0][0]),3)
y_solution = round(float(solution.args[0][1]),3)

# Make sure the window includes the solution
xmin = int(x_solution) - 20
xmax = int(x_solution) + 20
ymin = int(y_solution) - 20
ymax = int(y_solution) + 20
points = 2*(xmax-xmin)

# Define the x values once for the graph
graph_x = np.linspace(xmin,xmax,points)

# Define the y values for the graph
y1 = m1*graph_x + b1
y2 = m2*graph_x + b2

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

# line 1
plt.plot(graph_x, y1)

# line 2
plt.plot(graph_x, y2)

# point
plt.plot([x_solution],[y_solution],'ro')

plt.show()
print(" ")
print("Solution: (", x_solution, ",", y_solution, ")")


# Run this code and test it with different equations
```

```python
# Write your code here

import math
import random
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

class ProjectileGame:
    def __init__(self):
        # Game parameters
        self.gravity = 9.8  # m/s²
        self.wall_x = random.uniform(20, 80)
        self.wall_height = random.uniform(5, 15)
        self.max_height = 20

        # Rocket trajectory
        self.trajectory_x = []
        self.trajectory_y = []

        # Game state
        self.success = False
        self.attempts = 0

    def calculate_trajectory(self, initial_velocity, launch_angle):
        """Calculate the rocket's trajectory using quadratic functions"""
        # Convert angle to radians
        angle_rad = math.radians(launch_angle)

        # Initial velocity components
        vx = initial_velocity * math.cos(angle_rad)
        vy = initial_velocity * math.sin(angle_rad)

        # Time when rocket hits ground (y=0)
        time_of_flight = (2 * vy) / self.gravity

        # Generate trajectory points
        self.trajectory_x = []
        self.trajectory_y = []

        t_values = np.linspace(0, time_of_flight, 100)
        for t in t_values:
            x = vx * t
            y = vy * t - 0.5 * self.gravity * t**2
            if y >= 0:  # Only include points above ground
                self.trajectory_x.append(x)
                self.trajectory_y.append(y)

        return self.trajectory_x, self.trajectory_y

    def check_clearance(self):
        """Check if the rocket clears the wall"""
        if not self.trajectory_x:
            return False

        # Find rocket height at wall position
        wall_x = self.wall_x

        # Find the point in trajectory closest to wall_x
        closest_idx = min(range(len(self.trajectory_x)),
                         key=lambda i: abs(self.trajectory_x[i] - wall_x))

        rocket_height_at_wall = self.trajectory_y[closest_idx]

        return rocket_height_at_wall > self.wall_height

    def get_quadratic_equation(self, initial_velocity, launch_angle):
        """Return the quadratic equation for the trajectory"""
        angle_rad = math.radians(launch_angle)
        vx = initial_velocity * math.cos(angle_rad)
        vy = initial_velocity * math.sin(angle_rad)

        # The trajectory follows: y = (-g/(2vx²))x² + (vy/vx)x
        a = -self.gravity / (2 * vx**2)
        b = vy / vx
        c = 0  # Starting from (0,0)

        return a, b, c

    def plot_trajectory(self, initial_velocity, launch_angle):
        """Create an animated plot of the rocket's trajectory"""
        fig, ax = plt.subplots(figsize=(12, 8))

        # Set up the plot
        ax.set_xlim(0, 100)
        ax.set_ylim(0, self.max_height)
        ax.set_xlabel('Distance (m)')
        ax.set_ylabel('Height (m)')
        ax.set_title('Rocket Launch: Clear the Wall!')
        ax.grid(True, alpha=0.3)

        # Draw the wall
        ax.axvline(x=self.wall_x, ymin=0, ymax=self.wall_height/self.max_height,
                  color='red', linewidth=10, label=f'Wall ({self.wall_height:.1f}m high)')

        # Calculate trajectory
        x_points, y_points = self.calculate_trajectory(initial_velocity, launch_angle)

        # Create trajectory line and rocket point
        trajectory_line, = ax.plot([], [], 'b-', alpha=0.7, label='Trajectory')
        rocket_point, = ax.plot([], [], 'ro', markersize=10, label='Rocket')

        # Display quadratic equation
        a, b, c = self.get_quadratic_equation(initial_velocity, launch_angle)
        equation_text = ax.text(0.02, 0.95, '', transform=ax.transAxes, fontsize=12,
                               bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

        # Display result
        result_text = ax.text(0.02, 0.85, '', transform=ax.transAxes, fontsize=12,
                             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

        def animate(frame):
            if frame < len(x_points):
                # Update trajectory
                trajectory_line.set_data(x_points[:frame+1], y_points[:frame+1])
                rocket_point.set_data([x_points[frame]], [y_points[frame]])

                # Update equation (show full equation once we have some points)
                if frame == len(x_points) - 1:
                    equation_text.set_text(f'Quadratic Equation:\ny = {a:.4f}x² + {b:.4f}x')

                    # Check if rocket clears the wall
                    self.success = self.check_clearance()
                    if self.success:
                        result_text.set_text('SUCCESS! 🚀\nRocket cleared the wall!')
                        result_text.set_color('green')
                    else:
                        result_text.set_text('FAILED! 💥\nRocket hit the wall')
                        result_text.set_color('red')

            return trajectory_line, rocket_point, equation_text, result_text

        # Create animation
        anim = animation.FuncAnimation(fig, animate, frames=len(x_points)+10,
                                     interval=50, blit=False, repeat=False)

        ax.legend()
        plt.tight_layout()
        plt.show()

        return self.success

def play_game():
    """Main game function"""
    print("🎯 QUADRATIC FUNCTIONS PROJECTILE GAME 🎯")
    print("=" * 50)
    print("Launch a rocket from (0,0) to clear a wall!")
    print("The rocket's path follows a quadratic function:")
    print("y = (-g/(2vₓ²))x² + (vᵧ/vₓ)x")
    print("where vₓ = v₀·cos(θ), vᵧ = v₀·sin(θ)")
    print("=" * 50)

    # Create game instance
    game = ProjectileGame()

    print(f"\n📏 WALL INFORMATION:")
    print(f"   Position: {game.wall_x:.1f} meters away")
    print(f"   Height: {game.wall_height:.1f} meters")

    print(f"\n🎮 GAME PARAMETERS:")
    print(f"   Gravity: {game.gravity} m/s²")
    print(f"   Maximum display height: {game.max_height} meters")

    while True:
        game.attempts += 1
        print(f"\n--- Attempt {game.attempts} ---")

        try:
            # Get user input
            initial_velocity = float(input("Enter initial velocity (m/s, 15-40 recommended): "))
            launch_angle = float(input("Enter launch angle (degrees, 30-70 recommended): "))

            if initial_velocity <= 0 or launch_angle <= 0 or launch_angle >= 90:
                print("Please enter positive values for velocity and angle between 0-90 degrees.")
                continue

            # Show the animation
            print("\nLaunching rocket...")
            success = game.plot_trajectory(initial_velocity, launch_angle)

            if success:
                print(f"\n🎉 CONGRATULATIONS! You cleared the wall in {game.attempts} attempts!")

                # Show the quadratic equation
                a, b, c = game.get_quadratic_equation(initial_velocity, launch_angle)
                print(f"\n📊 The rocket's trajectory followed:")
                print(f"   y = {a:.4f}x² + {b:.4f}x")

                play_again = input("\nPlay again? (y/n): ").lower().strip()
                if play_again != 'y':
                    break
                else:
                    # Create new game with new random wall
                    game = ProjectileGame()
                    print(f"\n📏 NEW WALL: {game.wall_x:.1f}m away, {game.wall_height:.1f}m high")
            else:
                print("\n💥 The rocket didn't clear the wall. Try again!")
                print("Hint: Increase velocity or adjust the angle.")

        except ValueError:
            print("Please enter valid numbers.")
        except KeyboardInterrupt:
            print("\n\nThanks for playing!")
            break

def educational_explanation():
    """Provide educational explanation of quadratic functions in projectile motion"""
    print("\n" + "="*60)
    print("EDUCATIONAL EXPLANATION: Quadratic Functions in Projectile Motion")
    print("="*60)

    print("""
In projectile motion (ignoring air resistance), the path of an object follows a PARABOLA,
which is described by a QUADRATIC FUNCTION.

The general form of a quadratic function is:
    y = ax² + bx + c

For our rocket launching from (0,0):
• c = 0 (since we start at height 0)
• b = vᵧ/vₓ = tan(θ) (slope at launch)
• a = -g/(2vₓ²) (curvature due to gravity)

Where:
• v₀ = initial velocity
• θ = launch angle
• vₓ = v₀·cos(θ) = horizontal velocity component
• vᵧ = v₀·sin(θ) = vertical velocity component
• g = acceleration due to gravity (9.8 m/s²)

Key features of the parabola:
• Vertex: The highest point of the trajectory
• Roots: Where the projectile hits the ground (y=0)
• The 'a' coefficient determines how steep the parabola is

In the game, you're solving for v₀ and θ such that:
    y(wall_x) > wall_height
""")

# Run the game
if __name__ == "__main__":
    educational_explanation()
    play_game()

# This step does not have a test
```

```python
# Write your code here

import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, solve, Eq
import pandas as pd

def display_graph_and_table():
    """Display graph and table of values for any y= equation"""
    try:
        equation_str = input("Enter equation in terms of x (e.g., '2*x + 3', 'x**2 - 4', 'sin(x)'): ")

        # Define the function from user input
        def equation(x):
            return eval(equation_str)

        # Generate x values
        x_min = float(input("Enter minimum x value: "))
        x_max = float(input("Enter maximum x value: "))
        x_values = np.linspace(x_min, x_max, 100)
        y_values = equation(x_values)

        # Create table of values
        table_x = np.linspace(x_min, x_max, 20)
        table_y = equation(table_x)

        # Display table
        df = pd.DataFrame({'x': table_x, 'y': table_y})
        print("\nTable of Values:")
        print(df.round(3))

        # Plot graph
        plt.figure(figsize=(10, 6))
        plt.plot(x_values, y_values, 'b-', linewidth=2, label=f'y = {equation_str}')
        plt.grid(True, alpha=0.3)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title(f'Graph of y = {equation_str}')
        plt.legend()
        plt.show()

    except Exception as e:
        print(f"Error: {e}. Please check your equation format.")

def solve_system_two_equations():
    """Solve a system of two equations without graphing"""
    try:
        print("Enter first equation in terms of x and y (e.g., '2*x + 3*y - 6'): ")
        eq1_str = input()
        print("Enter second equation in terms of x and y (e.g., 'x - y + 1'): ")
        eq2_str = input()

        x, y = symbols('x y')

        # Create equations
        eq1 = Eq(eval(eq1_str), 0)
        eq2 = Eq(eval(eq2_str), 0)

        # Solve the system
        solution = solve((eq1, eq2), (x, y))

        if solution:
            print(f"\nSolution: x = {solution[x]:.3f}, y = {solution[y]:.3f}")
        else:
            print("No solution found or system is inconsistent")

    except Exception as e:
        print(f"Error: {e}. Please check your equation format.")

def graph_two_equations_intersection():
    """Graph two equations and plot the point of intersection"""
    try:
        print("Enter first equation in terms of x (e.g., '2*x + 3'): ")
        eq1_str = input()
        print("Enter second equation in terms of x (e.g., '-x + 5'): ")
        eq2_str = input()

        # Define functions
        def eq1(x):
            return eval(eq1_str)

        def eq2(x):
            return eval(eq2_str)

        # Generate x values
        x_min = float(input("Enter minimum x value: "))
        x_max = float(input("Enter maximum x value: "))
        x_values = np.linspace(x_min, x_max, 100)
        y1_values = eq1(x_values)
        y2_values = eq2(x_values)

        # Find intersection point
        x_sym = symbols('x')
        equation = Eq(eval(eq1_str), eval(eq2_str))
        intersection_x = solve(equation, x_sym)

        plt.figure(figsize=(10, 6))

        # Plot both equations
        plt.plot(x_values, y1_values, 'b-', linewidth=2, label=f'y = {eq1_str}')
        plt.plot(x_values, y2_values, 'r-', linewidth=2, label=f'y = {eq2_str}')

        # Plot intersection points
        if intersection_x:
            for x_val in intersection_x:
                if x_min <= float(x_val) <= x_max:
                    y_val = eq1(float(x_val))
                    plt.plot(float(x_val), y_val, 'go', markersize=8, label=f'Intersection ({float(x_val):.2f}, {y_val:.2f})')
                    print(f"Intersection point: ({float(x_val):.3f}, {y_val:.3f})")

        plt.grid(True, alpha=0.3)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Graph of Two Equations with Intersection Points')
        plt.legend()
        plt.show()

    except Exception as e:
        print(f"Error: {e}. Please check your equation format.")

def plot_quadratic_roots_vertex():
    """Given a, b, c in a quadratic equation, plot roots and vertex"""
    try:
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))

        # Calculate discriminant
        discriminant = b**2 - 4*a*c

        # Calculate vertex
        vertex_x = -b / (2*a)
        vertex_y = a*vertex_x**2 + b*vertex_x + c

        print(f"\nVertex: ({vertex_x:.3f}, {vertex_y:.3f})")

        # Calculate roots
        if discriminant > 0:
            root1 = (-b + np.sqrt(discriminant)) / (2*a)
            root2 = (-b - np.sqrt(discriminant)) / (2*a)
            print(f"Roots: {root1:.3f} and {root2:.3f}")
            roots = [root1, root2]
        elif discriminant == 0:
            root = -b / (2*a)
            print(f"Double root: {root:.3f}")
            roots = [root]
        else:
            real_part = -b / (2*a)
            imag_part = np.sqrt(-discriminant) / (2*a)
            print(f"Complex roots: {real_part:.3f} + {imag_part:.3f}i and {real_part:.3f} - {imag_part:.3f}i")
            roots = []

        # Generate x values for plotting
        x_range = max(5, abs(vertex_x) * 2)  # Dynamic range based on vertex
        x_min = vertex_x - x_range
        x_max = vertex_x + x_range
        x_values = np.linspace(x_min, x_max, 100)
        y_values = a*x_values**2 + b*x_values + c

        # Plot
        plt.figure(figsize=(10, 6))
        plt.plot(x_values, y_values, 'b-', linewidth=2, label=f'y = {a}x² + {b}x + {c}')

        # Plot vertex
        plt.plot(vertex_x, vertex_y, 'ro', markersize=8, label=f'Vertex ({vertex_x:.2f}, {vertex_y:.2f})')

        # Plot real roots
        for root in roots:
            if x_min <= root <= x_max:
                plt.plot(root, 0, 'go', markersize=8, label=f'Root ({root:.2f}, 0)')

        plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
        plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
        plt.grid(True, alpha=0.3)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Quadratic Equation with Roots and Vertex')
        plt.legend()
        plt.show()

    except Exception as e:
        print(f"Error: {e}. Please check your coefficients.")

def main_menu():
    """Display main menu and handle user choices"""
    while True:
        print("\n" + "="*50)
        print("MATHEMATICAL FUNCTIONS MENU")
        print("="*50)
        print("1. Display graph and table for y= equation")
        print("2. Solve system of two equations (no graphing)")
        print("3. Graph two equations and plot intersection")
        print("4. Plot quadratic roots and vertex")
        print("5. Exit")
        print("="*50)

        choice = input("Enter your choice (1-5): ").strip()

        if choice == '1':
            display_graph_and_table()
        elif choice == '2':
            solve_system_two_equations()
        elif choice == '3':
            graph_two_equations_intersection()
        elif choice == '4':
            plot_quadratic_roots_vertex()
        elif choice == '5':
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1-5.")

# Run the program
if __name__ == "__main__":
    main_menu()

# This step does not have a test
```

```python
# Write your code here

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
import matplotlib.widgets as widgets
from matplotlib.backend_bases import MouseButton
import sympy as sp

class GraphingCalculator:
    def __init__(self):
        self.fig, self.ax = plt.subplots(figsize=(12, 8))
        self.fig.suptitle('Interactive Graphing Calculator', fontsize=16)

        # Initialize variables
        self.functions = []
        self.x_range = [-10, 10]
        self.y_range = [-10, 10]
        self.zoom_factor = 1.2
        self.current_shade = None

        # Set up the plot
        self.setup_plot()

        # Connect event handlers
        self.fig.canvas.mpl_connect('button_press_event', self.on_click)
        self.fig.canvas.mpl_connect('key_press_event', self.on_key)

    def setup_plot(self):
        """Set up the initial plot appearance"""
        self.ax.clear()
        self.ax.set_xlim(self.x_range)
        self.ax.set_ylim(self.y_range)
        self.ax.axhline(y=0, color='k', linewidth=0.5)
        self.ax.axvline(x=0, color='k', linewidth=0.5)
        self.ax.grid(True, alpha=0.3)
        self.ax.set_xlabel('x')
        self.ax.set_ylabel('y')
        self.ax.set_title('Graph')

    def parse_function(self, func_str):
        """Parse a function string into a callable function"""
        try:
            # Replace common math notations
            func_str = func_str.replace('^', '**')

            # Create a lambda function
            x = sp.symbols('x')
            expr = sp.sympify(func_str)
            func = sp.lambdify(x, expr, modules=['numpy'])
            return func
        except:
            print(f"Error parsing function: {func_str}")
            return None

    def graph_function(self, func_str, color=None):
        """Graph a single function"""
        func = self.parse_function(func_str)
        if func is None:
            return

        x_vals = np.linspace(self.x_range[0], self.x_range[1], 1000)
        try:
            y_vals = func(x_vals)

            # Handle color assignment
            colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown']
            color = color or colors[len(self.functions) % len(colors)]

            self.ax.plot(x_vals, y_vals, label=func_str, color=color, linewidth=2)
            self.functions.append((func_str, func, color))

            self.ax.legend()
            plt.draw()
        except Exception as e:
            print(f"Error evaluating function: {e}")

    def create_table(self, func_str, start=-5, end=5, step=1):
        """Create a table of (x,y) values for a function"""
        func = self.parse_function(func_str)
        if func is None:
            return

        print(f"\nTable for: {func_str}")
        print("x\t|\ty")
        print("-" * 20)

        x_values = np.arange(start, end + step, step)
        for x in x_values:
            try:
                y = func(x)
                print(f"{x}\t|\t{y:.2f}")
            except:
                print(f"{x}\t|\tundefined")

    def shade_region(self, func_str, above=True, alpha=0.3):
        """Shade above or below a line"""
        func = self.parse_function(func_str)
        if func is None:
            return

        x_vals = np.linspace(self.x_range[0], self.x_range[1], 1000)
        y_vals = func(x_vals)

        # Create polygon for shading
        if above:
            vertices = list(zip(x_vals, y_vals)) + [(self.x_range[1], self.y_range[1]),
                                                   (self.x_range[0], self.y_range[1])]
        else:
            vertices = list(zip(x_vals, y_vals)) + [(self.x_range[1], self.y_range[0]),
                                                   (self.x_range[0], self.y_range[0])]

        polygon = Polygon(vertices, alpha=alpha, color='yellow')
        self.ax.add_patch(polygon)
        self.current_shade = (func_str, above)
        plt.draw()

    def solve_system(self, eq1_str, eq2_str):
        """Solve and graph a system of two equations"""
        # Parse equations
        func1 = self.parse_function(eq1_str)
        func2 = self.parse_function(eq2_str)

        if func1 is None or func2 is None:
            return

        # Graph both functions
        self.graph_function(eq1_str, 'blue')
        self.graph_function(eq2_str, 'red')

        # Find intersection points numerically
        x_vals = np.linspace(self.x_range[0], self.x_range[1], 1000)
        y1_vals = func1(x_vals)
        y2_vals = func2(x_vals)

        # Find where the difference changes sign
        diff = y1_vals - y2_vals
        intersections = []

        for i in range(len(diff) - 1):
            if diff[i] * diff[i + 1] <= 0:  # Sign change
                # Refine using binary search
                x_left, x_right = x_vals[i], x_vals[i + 1]
                for _ in range(10):  # 10 iterations for precision
                    x_mid = (x_left + x_right) / 2
                    y1_mid = func1(x_mid)
                    y2_mid = func2(x_mid)
                    diff_mid = y1_mid - y2_mid

                    if diff[i] * diff_mid <= 0:
                        x_right = x_mid
                    else:
                        x_left = x_mid

                x_intersect = (x_left + x_right) / 2
                y_intersect = func1(x_intersect)
                intersections.append((x_intersect, y_intersect))

        # Plot intersection points
        if intersections:
            for x, y in intersections:
                self.ax.plot(x, y, 'ro', markersize=8, label=f'Solution: ({x:.2f}, {y:.2f})')
            print(f"\nSolutions found: {intersections}")
        else:
            print("\nNo intersection points found in the current view")

        self.ax.legend()
        plt.draw()

    def zoom(self, direction='in'):
        """Zoom in or out on the graph"""
        x_center = (self.x_range[0] + self.x_range[1]) / 2
        y_center = (self.y_range[0] + self.y_range[1]) / 2

        x_width = self.x_range[1] - self.x_range[0]
        y_width = self.y_range[1] - self.y_range[0]

        if direction == 'in':
            self.x_range = [x_center - x_width / (2 * self.zoom_factor),
                           x_center + x_width / (2 * self.zoom_factor)]
            self.y_range = [y_center - y_width / (2 * self.zoom_factor),
                           y_center + y_width / (2 * self.zoom_factor)]
        else:  # zoom out
            self.x_range = [x_center - x_width * self.zoom_factor / 2,
                           x_center + x_width * self.zoom_factor / 2]
            self.y_range = [y_center - y_width * self.zoom_factor / 2,
                           y_center + y_width * self.zoom_factor / 2]

        self.setup_plot()
        self.redraw_functions()
        plt.draw()

    def solve_quadratic(self, a, b, c):
        """Solve a quadratic equation ax² + bx + c = 0"""
        discriminant = b**2 - 4*a*c

        print(f"\nSolving: {a}x² + {b}x + {c} = 0")
        print(f"Discriminant: {discriminant}")

        if discriminant > 0:
            root1 = (-b + np.sqrt(discriminant)) / (2*a)
            root2 = (-b - np.sqrt(discriminant)) / (2*a)
            print(f"Two real roots: x = {root1:.2f}, x = {root2:.2f}")

            # Graph the quadratic
            func_str = f"{a}*x**2 + {b}*x + {c}"
            self.graph_function(func_str, 'green')

            # Mark roots
            self.ax.plot(root1, 0, 'go', markersize=8, label=f'Root: {root1:.2f}')
            self.ax.plot(root2, 0, 'go', markersize=8, label=f'Root: {root2:.2f}')

        elif discriminant == 0:
            root = -b / (2*a)
            print(f"One real root: x = {root:.2f}")

            func_str = f"{a}*x**2 + {b}*x + {c}"
            self.graph_function(func_str, 'green')
            self.ax.plot(root, 0, 'go', markersize=8, label=f'Root: {root:.2f}')

        else:
            real_part = -b / (2*a)
            imaginary_part = np.sqrt(-discriminant) / (2*a)
            print(f"Two complex roots: {real_part:.2f} ± {imaginary_part:.2f}i")

            func_str = f"{a}*x**2 + {b}*x + {c}"
            self.graph_function(func_str, 'green')

        self.ax.legend()
        plt.draw()

    def redraw_functions(self):
        """Redraw all functions when zooming or resetting"""
        for func_str, func, color in self.functions:
            x_vals = np.linspace(self.x_range[0], self.x_range[1], 1000)
            try:
                y_vals = func(x_vals)
                self.ax.plot(x_vals, y_vals, label=func_str, color=color, linewidth=2)
            except:
                pass

        if self.current_shade:
            func_str, above = self.current_shade
            self.shade_region(func_str, above)

        self.ax.legend()

    def on_click(self, event):
        """Handle mouse clicks for interactive features"""
        if event.button == MouseButton.LEFT and event.inaxes == self.ax:
            print(f"Clicked at: ({event.xdata:.2f}, {event.ydata:.2f})")

    def on_key(self, event):
        """Handle keyboard shortcuts"""
        if event.key == 'i':
            self.zoom('in')
        elif event.key == 'o':
            self.zoom('out')
        elif event.key == 'r':  # reset view
            self.x_range = [-10, 10]
            self.y_range = [-10, 10]
            self.setup_plot()
            self.redraw_functions()
            plt.draw()
        elif event.key == 'c':  # clear all
            self.functions.clear()
            self.current_shade = None
            self.setup_plot()
            plt.draw()

    def show_instructions(self):
        """Display usage instructions"""
        print("\n" + "="*50)
        print("GRAPHING CALCULATOR INSTRUCTIONS")
        print("="*50)
        print("Keyboard Shortcuts:")
        print("  i - Zoom in")
        print("  o - Zoom out")
        print("  r - Reset view")
        print("  c - Clear all graphs")
        print("\nClick on the graph to see coordinates")
        print("="*50)

# Demonstration of the graphing calculator
def demo_calculator():
    calc = GraphingCalculator()
    calc.show_instructions()

    # Demo 1: Graph multiple functions
    print("\n1. Graphing multiple functions:")
    calc.graph_function("x**2")
    calc.graph_function("sin(x)")
    calc.graph_function("2*x + 1")

    # Demo 2: Create a table
    print("\n2. Creating a table:")
    calc.create_table("x**2", -3, 3, 1)

    # Demo 3: Shade region
    print("\n3. Shading above y = x")
    calc.shade_region("x", above=True)

    # Demo 4: Solve system of equations
    print("\n4. Solving system of equations:")
    calc.solve_system("x + 2", "-x + 4")

    # Demo 5: Solve quadratic
    print("\n5. Solving quadratic equation:")
    calc.solve_quadratic(1, -3, 2)  # x² - 3x + 2 = 0

    plt.tight_layout()
    plt.show()

# Interactive usage example
def interactive_calculator():
    calc = GraphingCalculator()
    calc.show_instructions()

    while True:
        print("\nChoose an option:")
        print("1. Graph function")
        print("2. Create table")
        print("3. Shade region")
        print("4. Solve system")
        print("5. Solve quadratic")
        print("6. Zoom in")
        print("7. Zoom out")
        print("8. Exit")

        choice = input("Enter your choice (1-8): ").strip()

        if choice == '1':
            func_str = input("Enter function (use x as variable, e.g., 'x**2 + 2*x + 1'): ")
            calc.graph_function(func_str)
        elif choice == '2':
            func_str = input("Enter function: ")
            start = float(input("Start x: ") or -5)
            end = float(input("End x: ") or 5)
            step = float(input("Step: ") or 1)
            calc.create_table(func_str, start, end, step)
        elif choice == '3':
            func_str = input("Enter function: ")
            direction = input("Shade (a)bove or (b)elow? ").lower()
            calc.shade_region(func_str, above=(direction == 'a'))
        elif choice == '4':
            eq1 = input("Enter first equation: ")
            eq2 = input("Enter second equation: ")
            calc.solve_system(eq1, eq2)
        elif choice == '5':
            a = float(input("Enter a: "))
            b = float(input("Enter b: "))
            c = float(input("Enter c: "))
            calc.solve_quadratic(a, b, c)
        elif choice == '6':
            calc.zoom('in')
        elif choice == '7':
            calc.zoom('out')
        elif choice == '8':
            break
        else:
            print("Invalid choice")

        plt.draw()

    plt.show()

# Run the demonstration
if __name__ == "__main__":
    print("Graphing Calculator Demo")
    demo_choice = input("Run (d)emo or (i)nteractive mode? ").lower()

    if demo_choice == 'd':
        demo_calculator()
    else:
        interactive_calculator()

# This step does not have a test
```
