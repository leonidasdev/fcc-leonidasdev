#Step 10 - Compound Interest

When you use percent increase formulas for money in the bank, it's called <i>compound interest</i>. The amount of money you earn beyond the original principle is the interest. # Step 11 - Continuous Growth

In the first formula, A = P(1 + r)<sup>t</sup>, money is compounded annually.<br>
In the second formula, A = P(1 + $\frac{r}{n})^{nt}$, money is compounded `n` times per year.<br>
As `n` gets to be a really big number, you get a different formula, A = Pe$^{rt}$, for continuous growth.<br>
In this formula, `e` is a constant, about equal to 2.718281828. The following code already prompts for the four variables once. Use those variables to compare the annuity from the three formulas. Notice you need to `import math` to use `math.e` (the best approximation of e).

```python
import math

p = float(input("Principle: "))
r = float(input("Rate: "))
t = int(input("Time: "))
n = int(input("N: "))

# Only change the following three formulas:
a_annual = p*(1+r)**t
a_n_times = p*(1+r/n)**(n*t)
a_continuous = p*math.e**(r*t) # use math.e in this formula

# Only change the code above this line

print("Compounded annually, anuity = ", a_annual)
print("Compounded ", n, "times per year, anuity = ", a_n_times)
print("Compounded continuously, anuity = ", a_continuous)


# Only change code above this line
import math_code_test_c as test
test.step11(In[-1].split('# Only change code above this line')[0])
```

# Step 22 - Functions for the Project

Define a function for calculating mortgage payments and a function for calculating investment balance. Use code you wrote in earlier steps. Each function should prompt for input and print the output.

```python
# Write your code here
def calculate_mortgage_payment():
    """
    Calculate monthly mortgage payment based on loan amount, interest rate, and loan term
    """
    print("\n=== Mortgage Payment Calculator ===")

    # Get user input
    principal = float(input("Enter the loan amount: $"))
    annual_rate = float(input("Enter the annual interest rate (%): "))
    years = int(input("Enter the loan term (years): "))

    # Convert annual rate to monthly and percentage to decimal
    monthly_rate = (annual_rate / 100) / 12

    # Convert years to months
    months = years * 12

    # Calculate monthly payment using the formula
    if monthly_rate > 0:
        monthly_payment = principal * (monthly_rate * (1 + monthly_rate) ** months) / ((1 + monthly_rate) ** months - 1)
    else:
        monthly_payment = principal / months

    # Display results
    print(f"\nMortgage Calculation Results:")
    print(f"Loan Amount: ${principal:,.2f}")
    print(f"Annual Interest Rate: {annual_rate}%")
    print(f"Loan Term: {years} years ({months} months)")
    print(f"Monthly Payment: ${monthly_payment:,.2f}")
    print(f"Total Payments: ${monthly_payment * months:,.2f}")
    print(f"Total Interest: ${(monthly_payment * months) - principal:,.2f}")

def calculate_investment_balance():
    """
    Calculate future investment balance based on initial deposit, monthly contributions, interest rate, and time
    """
    print("\n=== Investment Balance Calculator ===")

    # Get user input
    initial_deposit = float(input("Enter the initial deposit: $"))
    monthly_contribution = float(input("Enter the monthly contribution: $"))
    annual_rate = float(input("Enter the annual interest rate (%): "))
    years = int(input("Enter the investment period (years): "))

    # Convert annual rate to monthly and percentage to decimal
    monthly_rate = (annual_rate / 100) / 12

    # Convert years to months
    months = years * 12

    # Calculate future value
    # Future value of initial deposit
    future_value_initial = initial_deposit * (1 + monthly_rate) ** months

    # Future value of monthly contributions (annuity)
    if monthly_rate > 0:
        future_value_contributions = monthly_contribution * (((1 + monthly_rate) ** months - 1) / monthly_rate)
    else:
        future_value_contributions = monthly_contribution * months

    # Total future value
    total_future_value = future_value_initial + future_value_contributions

    # Calculate totals
    total_contributions = initial_deposit + (monthly_contribution * months)
    total_interest = total_future_value - total_contributions

    # Display results
    print(f"\nInvestment Calculation Results:")
    print(f"Initial Deposit: ${initial_deposit:,.2f}")
    print(f"Monthly Contribution: ${monthly_contribution:,.2f}")
    print(f"Annual Interest Rate: {annual_rate}%")
    print(f"Investment Period: {years} years ({months} months)")
    print(f"Total Contributions: ${total_contributions:,.2f}")
    print(f"Total Interest Earned: ${total_interest:,.2f}")
    print(f"Future Investment Balance: ${total_future_value:,.2f}")

# Main program to run the calculators
def main():
    while True:
        print("\n" + "="*50)
        print("FINANCIAL CALCULATOR")
        print("="*50)
        print("1. Calculate Mortgage Payment")
        print("2. Calculate Investment Balance")
        print("3. Exit")

        choice = input("\nEnter your choice (1-3): ")

        if choice == '1':
            calculate_mortgage_payment()
        elif choice == '2':
            calculate_investment_balance()
        elif choice == '3':
            print("Thank you for using the Financial Calculator!")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

        input("\nPress Enter to continue...")

# Run the program
if __name__ == "__main__":
    main()

# This step does not have a test
```

# Step 24 - New Function

Create a function to print the time required for money to double, given the rate. Use the continuous growth formula from an earlier step.

```python
# Write your code here

import math

def time_to_double():
    """
    Calculate and print the time required for money to double given a continuous interest rate
    Uses the continuous growth formula: A = P * e^(rt)
    """
    # Prompt for input
    rate = float(input("Enter the annual interest rate (as a percentage): "))

    # Convert percentage to decimal
    r = rate / 100

    # Continuous growth formula: A = P * e^(rt)
    # For doubling: 2P = P * e^(rt) → 2 = e^(rt) → ln(2) = rt → t = ln(2) / r
    time = math.log(2) / r

    # Print the result
    print(f"At {rate}% continuous interest, money will double in {time:.2f} years")

# This step does not have a test
```

```python
# Write your code here

import math
import numpy as np

class FinancialApp:
    def __init__(self):
        self.menu_options = {
            '1': 'Annuity Calculator',
            '2': 'Mortgage Payment Calculator',
            '3': 'Retirement Investment Calculator',
            '4': 'Doubling Time Calculator',
            '5': 'Logarithmic Equation Solver',
            '6': 'Scientific Notation Converter',
            '7': 'Exit'
        }

    def display_menu(self):
        print("\n" + "="*50)
        print("FINANCIAL CALCULATOR")
        print("="*50)
        for key, value in self.menu_options.items():
            print(f"{key}. {value}")
        print("="*50)

    def annuity_calculator(self):
        print("\n--- Annuity Calculator ---")
        print("Choose growth type:")
        print("1. Monthly compounding")
        print("2. Continuous growth")

        choice = input("Enter choice (1-2): ")

        if choice == '1':
            # Monthly compounding
            P = float(input("Enter principal amount: $"))
            r = float(input("Enter annual interest rate (%): ")) / 100
            n = int(input("Enter number of years: "))
            m = 12  # monthly compounding

            # Future value with monthly compounding
            A = P * (1 + r/m) ** (m*n)
            print(f"\nFuture Value with Monthly Compounding: ${A:,.2f}")
            print(f"Total Interest Earned: ${A - P:,.2f}")

        elif choice == '2':
            # Continuous growth
            P = float(input("Enter principal amount: $"))
            r = float(input("Enter annual interest rate (%): ")) / 100
            t = int(input("Enter number of years: "))

            # Future value with continuous growth
            A = P * math.exp(r * t)
            print(f"\nFuture Value with Continuous Growth: ${A:,.2f}")
            print(f"Total Interest Earned: ${A - P:,.2f}")
        else:
            print("Invalid choice!")

    def mortgage_calculator(self):
        print("\n--- Mortgage Payment Calculator ---")
        principal = float(input("Enter loan amount: $"))
        annual_rate = float(input("Enter annual interest rate (%): ")) / 100
        years = int(input("Enter loan term (years): "))

        monthly_rate = annual_rate / 12
        months = years * 12

        if monthly_rate > 0:
            monthly_payment = principal * (monthly_rate * (1 + monthly_rate) ** months) / ((1 + monthly_rate) ** months - 1)
        else:
            monthly_payment = principal / months

        print(f"\nMonthly Payment: ${monthly_payment:,.2f}")
        print(f"Total Payments: ${monthly_payment * months:,.2f}")
        print(f"Total Interest: ${(monthly_payment * months) - principal:,.2f}")

    def retirement_calculator(self):
        print("\n--- Retirement Investment Calculator ---")
        initial = float(input("Enter initial investment: $"))
        monthly = float(input("Enter monthly contribution: $"))
        annual_rate = float(input("Enter annual return rate (%): ")) / 100
        years = int(input("Enter years until retirement: "))

        monthly_rate = annual_rate / 12
        months = years * 12

        # Future value of initial investment
        future_initial = initial * (1 + monthly_rate) ** months

        # Future value of monthly contributions
        if monthly_rate > 0:
            future_contributions = monthly * (((1 + monthly_rate) ** months - 1) / monthly_rate)
        else:
            future_contributions = monthly * months

        total_future = future_initial + future_contributions
        total_contributions = initial + (monthly * months)

        print(f"\nFuture Retirement Balance: ${total_future:,.2f}")
        print(f"Total Contributions: ${total_contributions:,.2f}")
        print(f"Total Growth: ${total_future - total_contributions:,.2f}")

    def doubling_time_calculator(self):
        print("\n--- Doubling Time Calculator ---")
        rate = float(input("Enter annual interest rate (%): ")) / 100

        # Continuous compounding doubling time
        time = math.log(2) / rate

        # Rule of 72 approximation
        rule_72 = 72 / (rate * 100)

        print(f"\nExact doubling time: {time:.2f} years")
        print(f"Rule of 72 approximation: {rule_72:.1f} years")
        print(f"At {rate*100}% interest, money doubles every {time:.1f} years")

    def logarithmic_solver(self):
        print("\n--- Logarithmic Equation Solver ---")
        print("Solve equations of the form: log_b(x) = y")

        try:
            base = float(input("Enter base (b): "))
            result = float(input("Enter result (y): "))

            # Solve for x: log_b(x) = y → x = b^y
            x = base ** result

            print(f"Solution: x = {base}^{result} = {x}")
            print(f"Verification: log_{base}({x}) = {math.log(x) / math.log(base)}")

        except ValueError:
            print("Invalid input! Please enter numbers only.")

    def scientific_notation_converter(self):
        print("\n--- Scientific Notation Converter ---")
        print("1. Convert TO scientific notation")
        print("2. Convert FROM scientific notation")

        choice = input("Enter choice (1-2): ")

        if choice == '1':
            number = float(input("Enter number: "))
            # Convert to scientific notation
            sci_notation = f"{number:.2e}"
            print(f"Scientific notation: {sci_notation}")

        elif choice == '2':
            sci_input = input("Enter number in scientific notation (e.g., 1.23e4): ")
            try:
                # Convert from scientific notation
                number = float(sci_input)
                print(f"Decimal form: {number:,.2f}")
            except ValueError:
                print("Invalid scientific notation format!")
        else:
            print("Invalid choice!")

    def run(self):
        print("Welcome to the Financial Calculator App!")

        while True:
            self.display_menu()
            choice = input("\nEnter your choice (1-7): ")

            if choice == '1':
                self.annuity_calculator()
            elif choice == '2':
                self.mortgage_calculator()
            elif choice == '3':
                self.retirement_calculator()
            elif choice == '4':
                self.doubling_time_calculator()
            elif choice == '5':
                self.logarithmic_solver()
            elif choice == '6':
                self.scientific_notation_converter()
            elif choice == '7':
                print("Thank you for using the Financial Calculator!")
                break
            else:
                print("Invalid choice! Please enter 1-7.")

            input("\nPress Enter to continue...")

# Create and run the financial app
if __name__ == "__main__":
    app = FinancialApp()
    app.run()1

# This step does not have a test
```
